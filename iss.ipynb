```python
import requests
import time
import matplotlib as mlp
import matplotlib.pyplot as plt

link = "https://api.wheretheiss.at/v1/satellites/25544"

longitudes = []
latitudes = []
timestamps = []

try_amount = 101
i = 0

while i < try_amount:
    req = requests.get(link)
    data =  req.json()
    tist = data["timestamp"]
    lats = data["latitude"]
    longs = data["longitude"]

    longitudes.append(longs)
    latitudes.append(lats)
    timestamps.append(tist)

    i += 1
    print(i, tist, lats, longs)
    time.sleep(10)

plt.plot(longitudes, latitudes, marker="o", markersize=3, linestyle="-")
plt.title("ISS Path")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
